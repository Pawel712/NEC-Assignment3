```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
```

```python
with open('../Datasets/A3-dataset1Modified.csv', 'r') as file:
   dataset = pd.read_csv('../Datasets/A3-dataset1Modified.csv', sep=',')

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

#Euclidean distance matrix
distance_matrix = pdist(X, metric='euclidean')

linkage_matrix_upgma = linkage(distance_matrix, method='average')

linkage_matrix_complete = linkage(distance_matrix, method='complete')

#dendrogram for UPGMA linkage
plt.figure(figsize=(15, 10))
dendrogram(linkage_matrix_upgma, labels=y, 
            color_threshold=3)
plt.title("Dendrogram (UPGMA linkage)")
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()


#dendrogram for Complete Linkage
plt.figure(figsize=(15, 10))
dendrogram(linkage_matrix_complete, labels=y, 
            color_threshold=5)
plt.title("Dendrogram (Complete Linkage)")
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()

```

```python
import numpy as np
# Unique values in y
unique, counts = np.unique(y, return_counts=True)

print("unique values in y: ", dict(zip(unique, counts)))
```
